**Replacing outliers with median**

```python
#1
def handle_outliers(x, column):
    Q1 = x[column].quantile(0.25)
    Q3 = x[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    x[column] = x[column].apply(lambda l: x[column].median() if l < lower_bound or l > upper_bound else l)
    return x
for column in x.columns:
  x=handle_outliers(x,column)
```

```python
x.plot(kind='box',figsize=(17,17),layout=(4,4),sharex=False,subplots=True)
```

```python
# x.plot(kind='box',figsize=(17,17),layout=(4,4),sharex=False,subplots=True)
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
X_train = x_train_smote
Y_train = y_train_smote
y_test_copy=y_test

print("X_train shape:", X_train.shape)
print("X_test shape:", x_test.shape)
print("Y_train shape:", Y_train.shape)
print("y_test shape:", y_test_copy.shape)
```

```python
Y_train.value_counts()
```

```python
y_test_new = [int(label)-1 for label in y_test_copy]
Y_train_new = [ int(label)-1   for label in Y_train ]
```

```python
Y_train_new = np.array(Y_train_new)
y_test_new = np.array(y_test_new)
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
Y_train_new
```

```python
y_test_new
```

```python
num_classes=3
Y_train=to_categorical(Y_train_new, num_classes)
y_test_cnn=to_categorical(y_test_new, num_classes)
Y_train[:3]
```

```python
Y_train_new.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, GRU, Dense , Flatten
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, GRU, Dense, Flatten , Reshape
from tensorflow.keras.layers import LeakyReLU
from tensorflow.keras.optimizers import Adam, SGD
from tensorflow.keras.optimizers import Adam
from keras.regularizers import l2
from keras.regularizers import l1
# Assuming the input data has the shape (num_samples, num_timesteps, num_features)
# In your case, num_timesteps will be 1 as you have 5 features for each sample.


num_timesteps = X_train.shape[1]




# Assuming the input data has the shape (num_samples, num_timesteps, num_features)
# In your case, num_timesteps will be 1 as you have 5 features for each sample.

models = Sequential([
    Conv1D(32, 3,  activation=LeakyReLU(alpha=0.01)   , kernel_regularizer=l1(0.01) , padding='same',     input_shape=(8,1)),  # (num_samples, num_timesteps=1, num_features=5)
    MaxPooling1D(2, strides=2),
    Conv1D(64, 3, activation=LeakyReLU(alpha=0.01), padding='same'),
    MaxPooling1D(2, strides=2),
    Flatten(),  # Flatten the output of the Conv1D layers
       Reshape((-1, 64)),
    GRU(64  , dropout=0.4),
    Dense(64, activation=LeakyReLU(alpha=0.01)),
    Dense(3, activation='softmax')  # Output layer with 3 target labels
])


models.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])


models.summary()
```

```python

# Train the model
# batch_size = 64
# epochs = 20

statics=models.fit(X_train, Y_train, batch_size=64, epochs=20 , validation_split=0.2)
```

```python
# Evaluate the model on the test set
loss, accuracy_cnn = models.evaluate(x_test, y_test_cnn)
print("Test Loss:", loss)
print("Test Accuracy:", accuracy_cnn)
```

```python
y_pred_cnn=models.predict(x_test)
y_pred_cnn
```
